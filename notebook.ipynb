### Data Engineering

```python
import ast
def convert(convObj):
    L = []
    for i in ast.literal_eval(convObj):
        L.append(i['name'])
    return L
```

```python
def convertDirector(convObj):
    L = []
    for i in ast.literal_eval(convObj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
# data_movie
data_movies.genres = data_movies.genres.apply(convert)
data_movies.keywords = data_movies.keywords.apply(convert)
data_movies.production_companies = data_movies.production_companies.apply(convert)
data_movies.production_countries = data_movies.production_countries.apply(convert)
data_movies.spoken_languages = data_movies.spoken_languages.apply(convert)

# data_credit

data_credit.cast = data_credit.cast.apply(convert)
data_credit.crew = data_credit.crew.apply(convertDirector)
```

```python
data_movies.overview = data_movies.overview.apply(lambda x: x.split(' '))
```

```python
data_movies.overview = data_movies.overview.apply(lambda x: [i.replace(' ', '') for i in x])
data_movies.keywords = data_movies.keywords.apply(lambda x: [i.replace(' ', '') for i in x])
data_movies.genres = data_movies.genres.apply(lambda x: [i.replace(' ', '') for i in x])
data_credit.cast = data_credit.cast.apply(lambda x: [i.replace(' ', '') for i in x])
data_credit.crew = data_credit.crew.apply(lambda x: [i.replace(' ', '') for i in x])
```

```python
data_credit = data_credit.drop('title',axis=1)
movie = pd.merge(data_movies, data_credit, left_on=['id'], right_on=['movie_id'])
```

```python
movie
```

```python
movie['tags'] = movie['overview']+ movie['keywords'] + movie['genres'] + movie['cast'] + movie['crew']
```

```python
new_df = movie[['id','original_title','tags']]
new_df
```

```python
new_df.tags = new_df.tags.apply(lambda x: " ".join(x))
new_df.tags = new_df.tags.apply(lambda x: x.lower())
```

```python
new_df
```

```python
from nltk.stem.porter import PorterStemmer
from wordcloud import STOPWORDS
import re
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    text = [x for x in text.split(' ') if x not in STOPWORDS]
    text = ' '.join(text)
    text = re.sub('[^a-z]', ' ', text)
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df.tags.apply(stem)
```

```python
new_df.tags = new_df.tags.apply(stem)
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000)
```

```python
vectors = cv.fit_transform(new_df.tags).toarray()
```

```python
sorted(cv.vocabulary_.items(),key=lambda x: x[1],reverse=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0][1:10]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['original_title']==movie].index[0]
    distancess = similarity[movie_index]
    movie_list = sorted(list(enumerate(distancess)),key=lambda x: x[1],reverse=True)[1:10]
    
    recommended_movies = []
    for i in movie_list:
        recommended_movies.append([new_df.iloc[i[0]].original_title])
        # print(movies.iloc[i[0]].original_title)
        # print([i[1]])
        
    return recommended_movies
```

```python
recommend('Batman Begins')
```
